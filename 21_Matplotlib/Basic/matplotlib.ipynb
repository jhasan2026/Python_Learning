```python
plt.plot(batsman['index'],batsman['V Kohli'],color="green")
plt.plot(batsman['index'],batsman['RG Sharma'],color="red")

plt.title("Rohit Sharma VS Virat Kohli")
plt.xlabel("Season")
plt.ylabel("Run Scored")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle="dashed",linewidth=3,marker='D',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color="red",linestyle='dotted',linewidth=2,marker='.',markersize=10,label="Rohit")

plt.title("Rohit Sharma VS Virat Kohli")
plt.xlabel("Season")
plt.ylabel("Run Scored")
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

# Bar Chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children)
```

```python
plt.barh(colors,children,)
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
# multiple bar
plt.bar(np.arange(df.shape[0]) - 0.2 , df['2015'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0])  , df['2016'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0]) + 0.2 , df['2017'],width=0.2,color='yellow')
plt.xticks(np.arange(df.shape[0]) , df['batsman'])
plt.show()
```

```python
# xtics
children = [10,20,40,10,30]
colors = ['red red red','blue blue blue','green','yellow','pink']

plt.bar(colors,children,color='#000')
plt.xticks(rotation='vertical')
```

```python
## stack bar chart

plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')
plt.legend()
plt.show()
```
